```python
def plot(data: pd.DataFrame, stream_num: list, 
         start_seq: int, end_seq: int, ls = '--', marker=False):

    n_streams = len(stream_num)
    fig, axes = plt.subplots(n_streams, 1, figsize=(20, n_streams*4))

    plot_data = data.loc[start_seq:end_seq, :]
    for i, stream in enumerate(stream_num):
        axes[i].plot(plot_data.index, plot_data.iloc[:,  stream], 
                     color='grey', ls=ls, lw=1.5, zorder=1)

        anomaly = [0, 1]
        sizes   = [20, 20]
        markers = ['o', 'D']
        colors  = ['green', 'red']

        if marker == False:
            markers = ['', '']

        for a, s, m, c in zip(anomaly, sizes, markers, colors):
            axes[i].scatter(plot_data.loc[plot_data.anomaly == a, :].index, 
                            plot_data.loc[plot_data.anomaly == a, :].iloc[:, stream].values, 
                            marker=m, s=s, color=c, zorder = 2)

        axes[i].set_xlabel('Time', fontsize = 15)
        axes[i].set_ylabel('Stream â„–{}'.format(stream), fontsize = 15)
        axes[i].grid(color='blue', which='major', linestyle=':',
                     linewidth=0.5)
```

```python
class NominalDataset(Dataset):

	def __init__(self, x: np.array, seq_length: int, step: int):
		super(NominalDataset, self).__init__()

		if not isinstance(x, np.ndarray):
			raise Exception("X isn't numpy.ndarray ")

		self.n_samples = x.shape[0] - step * (seq_length - 1) - 1
		self.data = [
			(
				torch.from_numpy(x[i:i+(seq_length*step):step, :].astype(np.float32)),
				torch.from_numpy(x[i+((seq_length-1)*step)+1, :].astype(np.float32))
			)

			for i in range(self.n_samples)
		]
	

	def __getitem__(self, index: int) -> tuple:
		return self.data[index][0], self.data[index][1]


	def __len__(self) -> int:
		return self.n_samples
```

```python
class Encoder(nn.Module):

	def __init__(self, input_size: int, hidden_size: int, 
				num_layers: int, seq_length: int, z_size: int):
		super(Encoder, self).__init__()

		self.z_size = z_size
		self.num_layers = num_layers
		self.hidden_size = hidden_size

		self.lstm = nn.LSTM(input_size, hidden_size, num_layers, 
							batch_first=True, bidirectional=True)

		# mu
		self.m_layers = nn.ModuleList([

			nn.Sequential(

				nn.Linear(2*hidden_size, hidden_size),
				nn.LeakyReLU(0.1),
				nn.Linear(hidden_size, z_size)

			)

			for _ in range(seq_length)
		])

		# logvar
		self.v_layers = nn.ModuleList([ 

			nn.Sequential(

				nn.Linear(2*hidden_size, hidden_size),
				nn.LeakyReLU(0.1),
				nn.Linear(hidden_size, z_size)

			)

			for _ in range(seq_length)
			])


	def forward(self, x: torch.tensor) -> torch.tensor:

		h0 = torch.zeros(2 * self.num_layers, x.size(0), self.hidden_size)
		c0 = torch.zeros(2 * self.num_layers, x.size(0), self.hidden_size)

		output, _ = self.lstm(x, (h0, c0))
		z = torch.zeros(x.size(0), x.size(1), self.z_size)

		mu = torch.zeros(x.size(0), x.size(1), self.z_size)
		logvar = torch.zeros(x.size(0), x.size(1), self.z_size)

		for indx, (m, v) in enumerate(zip(self.m_layers, self.v_layers)):
			mu[:, indx, :]  = m(output[:, indx, :])
			logvar[:, indx, :] = F.softplus(v(output[:, indx, :]))

			# reparametrazation trick
			std = torch.exp(logvar[:, indx, :] / 2)
			eps = torch.randn_like(std)

			z[:, indx, :] = mu[:, indx, :] + eps*std

		return z, mu, logvar
```

```python
class Decoder(nn.Module):

	def __init__(self, z_size:  int, hidden_size: int, 
				num_layers: int, seq_length: int, output_size: int):
		super(Decoder, self).__init__()

		self.num_layers = num_layers
		self.output_size = output_size
		self.hidden_size = hidden_size

		self.lstm = nn.LSTM(z_size, hidden_size, num_layers, 
							batch_first=True, bidirectional=True)


		self.layers = nn.ModuleList([
			nn.Sequential(

				nn.Linear(hidden_size * 2, hidden_size),
				nn.LeakyReLU(0.2)

			)

			for _ in range(seq_length)

		])


		self.h2 = output_size * 2
		self.h4 = output_size * 4

		self.m = nn.Sequential(

			nn.Flatten(),
			nn.Linear(
				seq_length * hidden_size, 
				seq_length * hidden_size, bias=False
			),
			nn.LeakyReLU(0.2),
			nn.BatchNorm1d(seq_length * hidden_size),

			nn.Linear(
				seq_length * hidden_size, 
				seq_length * self.h4
			),
			nn.LeakyReLU(0.2),
		)


		# mu
		self.m_layers = nn.ModuleList([
			nn.Linear(self.h4, output_size) for _ in range(seq_length)
		])

		# logvar
		self.v_layers = nn.ModuleList([
			nn.Linear(self.h4, output_size) for _ in range(seq_length)
		])


	def forward(self, z: torch.tensor) -> torch.tensor:

		h0 = torch.zeros(2*self.num_layers, z.size(0), self.hidden_size)
		c0 = torch.zeros(2*self.num_layers, z.size(0), self.hidden_size)

		output, _ = self.lstm(z, (h0, c0))
		hidden1 = torch.zeros(z.size(0), z.size(1), self.hidden_size)

		for indx, m in enumerate(self.layers):
			hidden1[:, indx, :] = m(output[:, indx, :])

		hidden2 = self.m(hidden1)
		hidden2 = hidden2.view(z.size(0), z.size(1), -1)

		mu = torch.zeros(z.size(0), z.size(1), self.output_size)
		logvar = torch.zeros(z.size(0), z.size(1), self.output_size)


		for indx, (m, v) in enumerate(zip(self.m_layers, self.v_layers)):
			mu[:, indx, :]  = m(hidden2[:, indx, :])
			logvar[:, indx, :] = F.softplus(v(hidden2[:, indx, :]))

		return mu, logvar
```

```python
class VAE(nn.Module):

	def __init__(self, encoder, decoder):
		super(VAE, self).__init__()

		self.enc = encoder
		self.dec = decoder

	def forward(self, x: torch.tensor) -> tuple:

		z, z_mu, z_logvar = self.enc(x)
		x_mu, x_logvar = self.dec(z)

		return x_mu, x_logvar, z_mu, z_logvar
```

```python
# Hyperparameters

INPUT_SIZE  = 8
OUTPUT_SIZE = 8
Z_SIZE = 3

STEP = 1
SEQ_LENGTH = 100
BATCH_SIZE = 64
```

```python
# Dataset
dataset = NominalDataset(train_data, seq_length=SEQ_LENGTH, step=STEP)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
# part of VAE
encoder = Encoder(
    
    input_size=INPUT_SIZE, 
    hidden_size=80, 
    num_layers=1, 
    seq_length=SEQ_LENGTH, 
    z_size=Z_SIZE

)

decoder = Decoder(
    
    z_size=Z_SIZE, 
    hidden_size=80, 
    num_layers=1, 
    seq_length=SEQ_LENGTH, 
    output_size=OUTPUT_SIZE

)

model = VAE(encoder, decoder)
```

```python
def vae_loss(x_mu, x_logvar, x, mu, logvar, landa = 1):
    
    likelihood = torch.mean(0.5 * (x - x_mu) ** 2 / torch.exp(x_logvar) + \
                            torch.log(torch.sqrt(2*np.pi*torch.exp(x_logvar))), dim=0)
    
    KLD = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp(), dim=0)
    
    return torch.sum(likelihood), torch.sum(KLD)
```

```python
optimizer = optim.Adam(model.parameters(), lr=1e-3)
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[12,17], gamma=0.1)
```

```python
LANDA = 1
for epoch in range(10):
    
    mu_values  = torch.tensor([])
    var_values = torch.tensor([])    
        
    KLDLosses = torch.tensor([])
    likelihoodLosses = torch.tensor([])

    
    for sample, target in dataloader:
        
        x_mu, x_logvar, z_mu, z_logvar = model(sample)

        mu_values   = torch.cat((mu_values,  torch.mean(z_mu, dim=2)), dim=0)
        var_values  = torch.cat((var_values, torch.mean(z_logvar.exp(), dim=2)), dim=0)
        
    
        #VAE train
        
        likelihood, KLD = vae_loss(x_mu, x_logvar, sample, z_mu, z_logvar, LANDA)
        loss = likelihood + KLD

        KLDLosses = torch.cat(
            (KLDLosses, torch.unsqueeze(KLD, 0)), dim=0
        )
        
        likelihoodLosses = torch.cat(
            (likelihoodLosses, torch.unsqueeze(likelihood, 0)), dim=0
        )

        model.zero_grad()
        loss.backward()

        # cliping
        torch.nn.utils.clip_grad_norm_(
            model.parameters(), max_norm=1, norm_type=2
        )

        optimizer.step()
    
    scheduler.step()
    print("""    \nEpoch {}: LikeliHood: {:.5f}, KLD: {:.5f}
    Mean: [{}\n\t   {}] 
    Var:  [{}\n\t   {}]""".format(epoch, torch.mean(likelihoodLosses), torch.mean(KLDLosses),
                                  torch.mean(mu_values, dim=0)[:5].detach(), torch.mean(mu_values, dim=0)[-5:].detach(),
                                  torch.mean(var_values, dim=0)[:5].detach(), torch.mean(var_values, dim=0)[-5:].detach()
    ))
```

```python
from torch.distributions.multivariate_normal import MultivariateNormal
```

```python
# Dataset
dataset = NominalDataset(test_data, seq_length=SEQ_LENGTH, step=STEP)
dataloader = DataLoader(dataset, batch_size=256, shuffle=False)
```

```python
mz = MultivariateNormal(torch.zeros(Z_SIZE), torch.eye(Z_SIZE))
```

```python
model.eval()

probx_data   = np.empty(test_data.shape[0])
probz_data   = np.empty(test_data.shape[0])
stream_data = np.empty(test_data.shape)

for i, (sample, _) in enumerate(dataloader):
    x_mu, x_logvar, z_mu, z_logvar = model(sample)
    
    n = sample.size(0)
    for j in range(n):
        index = i * n + j

        mx = MultivariateNormal(
            x_mu[j, -1, :], 
            torch.diag(torch.exp(x_logvar[j, -1, :]))
        )
        
        eps = torch.randn_like(z_mu[j, -1, :])
        z = z_mu[j, -1, :] + eps * torch.exp(z_logvar[j, -1, :] / 2)
    
        
        probx_data[index]   = mx.log_prob(sample[j, -1, :]).detach().numpy()
        probz_data[index]   = mz.log_prob(z).detach().numpy()
        stream_data[index, :] = x_mu[j, -1, :].detach().numpy()
        #stream_data[index, :] = mx.sample().detach().numpy()

model.train()
```
