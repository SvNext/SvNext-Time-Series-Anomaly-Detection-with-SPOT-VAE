```python
def plot(data: pd.DataFrame, stream_num: list, 
         start_seq: int, end_seq: int, ls = '--', marker=False):

    n_streams = len(stream_num)
    fig, axes = plt.subplots(n_streams, 1, figsize=(20, n_streams*4))

    plot_data = data.loc[start_seq:end_seq, :]
    for i, stream in enumerate(stream_num):
        axes[i].plot(plot_data.index, plot_data.iloc[:,  stream], 
                     color='grey', ls=ls, lw=1.5, zorder=1)

        anomaly = [0, 1]
        sizes   = [20, 20]
        markers = ['o', 'D']
        colors  = ['green', 'red']

        if marker == False:
            markers = ['', '']

        for a, s, m, c in zip(anomaly, sizes, markers, colors):
            axes[i].scatter(plot_data.loc[plot_data.anomaly == a, :].index, 
                            plot_data.loc[plot_data.anomaly == a, :].iloc[:, stream].values, 
                            marker=m, s=s, color=c, zorder = 2)

        axes[i].set_xlabel('Time', fontsize = 15)
        axes[i].set_ylabel('Stream â„–{}'.format(stream), fontsize = 15)
        axes[i].grid(color='blue', which='major', linestyle=':',
                     linewidth=0.5)
```

```python
from vae import VAE
from encoder import Encoder
from decoder import Decoder
from dataset import NominalDataset
```

```python
# Hyperparameters

INPUT_SIZE  = train_data.shape[1]
OUTPUT_SIZE = train_data.shape[1]
Z_SIZE = 3

STEP = 1
BATCH_SIZE = 64
```

```python
# Dataset
dataset = NominalDataset(train_data, seq_length=SEQ_LENGTH, step=STEP)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
# part of VAE
encoder = Encoder(
    
    input_size=INPUT_SIZE, 
    hidden_size=80, 
    num_layers=1, 
    seq_length=SEQ_LENGTH, 
    z_size=Z_SIZE

)

decoder = Decoder(
    
    z_size=Z_SIZE, 
    hidden_size=80, 
    num_layers=1, 
    seq_length=SEQ_LENGTH, 
    output_size=OUTPUT_SIZE

)

model = VAE(encoder, decoder)
```

```python
def vae_loss(x_mu, x_logvar, x, mu, logvar, landa = 1):
    
    likelihood = torch.mean(0.5 * (x - x_mu) ** 2 / torch.exp(x_logvar) + \
                            torch.log(torch.sqrt(2*np.pi*torch.exp(x_logvar))), dim=0)
    
    KLD = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp(), dim=0)
    
    return torch.sum(likelihood), torch.sum(KLD)
```

```python
optimizer = optim.Adam(model.parameters(), lr=1e-4)
scheduler = optim.lr_scheduler.MultiStepLR(optimizer, milestones=[5,7], gamma=0.1)
```

```python
LANDA = 1
for epoch in range(10):
    
    mu_values  = torch.tensor([])
    var_values = torch.tensor([])    
        
    KLDLosses = torch.tensor([])
    likelihoodLosses = torch.tensor([])

    
    for sample, target in dataloader:
        
        x_mu, x_logvar, z_mu, z_logvar = model(sample)

        mu_values   = torch.cat((mu_values,  torch.mean(z_mu, dim=2)), dim=0)
        var_values  = torch.cat((var_values, torch.mean(z_logvar.exp(), dim=2)), dim=0)
        
    
        #VAE train
        
        likelihood, KLD = vae_loss(x_mu, x_logvar, sample, z_mu, z_logvar, LANDA)
        loss = likelihood + KLD

        KLDLosses = torch.cat(
            (KLDLosses, torch.unsqueeze(KLD, 0)), dim=0
        )
        
        likelihoodLosses = torch.cat(
            (likelihoodLosses, torch.unsqueeze(likelihood, 0)), dim=0
        )

        model.zero_grad()
        loss.backward()

        # cliping
        torch.nn.utils.clip_grad_norm_(
            model.parameters(), max_norm=1, norm_type=2
        )

        optimizer.step()
    
    scheduler.step()
    print("""    \nEpoch {}: LikeliHood: {:.5f}, KLD: {:.5f}
    Mean: [{}\n\t   {}] 
    Var:  [{}\n\t   {}]""".format(epoch, torch.mean(likelihoodLosses), torch.mean(KLDLosses),
                                  torch.mean(mu_values, dim=0)[:5].detach(), torch.mean(mu_values, dim=0)[-5:].detach(),
                                  torch.mean(var_values, dim=0)[:5].detach(), torch.mean(var_values, dim=0)[-5:].detach()
    ))
```

```python
cnames = results.columns.drop('anomaly')
```

```python
order_cnames = ['anomaly', 'log_prob_x', 'log_prob_z']
for cname in cnames:
    
    c1 = 'mu_{}'.format(cname)
    c2 = 'std_{}'.format(cname)
    
    results[c1], results[c2] = 0, 0
    order_cnames.extend([cname, c1, c2])
    
results['log_prob_x'] = 0
results['log_prob_z'] = 0
results = results[order_cnames]
```

```python
test_batch_size = 64
dataset = NominalDataset(test_data, seq_length=SEQ_LENGTH, step=STEP)
dataloader = DataLoader(dataset, batch_size=test_batch_size, shuffle=False)
```

```python
model.eval()
mz = MultivariateNormal(
    torch.zeros(Z_SIZE), torch.eye(Z_SIZE)
)

for i, (sample, _) in enumerate(dataloader):
    x_mu, x_logvar, z_mu, z_logvar = model(sample)
    
    n = sample.size(0)
    for j in range(n):
        index = i * n + j

        mx = MultivariateNormal(
            x_mu[j, -1, :], 
            torch.diag(torch.exp(x_logvar[j, -1, :]))
        )
        
        # think
        eps = torch.randn_like(z_mu[j, -1, :])
        z = z_mu[j, -1, :] + (torch.sign(eps) - eps) * torch.exp(z_logvar[j, -1, :] / 2)
        
        for ind, cname in enumerate(cnames):
            c1, c2 = 'mu_{}'.format(cname), 'std_{}'.format(cname)
            
            results.loc[index, c1] = x_mu[j, -1, ind].item()
            results.loc[index, c2] = torch.exp(x_logvar[j, -1, ind] / 2).item()
        
        
        results.loc[index, 'log_prob_z'] = mz.log_prob(z).item()
        results.loc[index, 'log_prob_x'] = mx.log_prob(sample[j, -1, :]).item()

model.train()
```

```python
s = biSPOT(q)
s.fit(spot_init, spot_data)
s.initialize()
```

```python
generator = s.run()
for _ in generator:
  pass
```

```python
plot_data = results.loc[n_init:, ['log_prob_z', 'anomaly']]
plot_data.reset_index(drop=True, inplace=True)

plot_data['thresholds'] = s.markers['lower_thresholds']
condition = plot_data.thresholds > plot_data.log_prob_z

plot_data['pred_anomaly'] = 0
plot_data.loc[condition, 'pred_anomaly'] = 1
```
